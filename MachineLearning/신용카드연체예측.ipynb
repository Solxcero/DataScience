```python
import pandas as pd
df = pd.read_csv('./UCI_Credit_Card.csv',encoding = 'utf-8')
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.rename(columns={'PAY_0':'PAY_1','default.payment.next.month':'default'}, inplace=True)
df.head()
```

```python
y_target = df['default']
X_features = df.drop('default',axis=1)
y_target.value_counts()
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corr = X_features.corr()
corr
```

```python
plt.figure(figsize=(18,18))
sns.heatmap(corr, annot=True, fmt='.1g',cmap='Purples')
# 다중공선성 발생 보여줌 --> 상관도가 높은 피처들 간에는 PCA 효율이 좋음 
# feature-selection, Lasso, Ridge, RandomForest
```

```python
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

```python
#BILL_AMT1~BILL_AMT6 까지 6개의 속성명 생성
cols_bill = ['BILL_AMT' +str(i) for i in range(1,7)]
print('대상 속성명:',cols_bill)

#2개의 PCA 속성을 가진 PCA 객체 생성하고, explained_variance_ratio_ 계산 위해 fit() 호출
scaler = StandardScaler()
df_cols_scaled = scaler.fit_transform(X_features[cols_bill])
pca1 = PCA(n_components=1)
pca1.fit(df_cols_scaled)
PCA1= pca1.transform(df_cols_scaled)

print('PCA Component별 변동성:', pca1.explained_variance_ratio_)

# 6개의 피처를 2개의 피처로 PCA 변환했을 때 첫번째 컴포넌트가 전체 변동성의 90% 설명
```

```python
print(PCA1.shape)
```

```python
X_features['PCA1'] = PCA1
X_features.drop(cols_bill,axis=1,inplace=True)
X_features.head()
```

```python
#BILL_AMT1~BILL_AMT6 까지 6개의 속성명 생성
cols_pay = ['PAY_' +str(i) for i in range(1,7)]
print('대상 속성명:',cols_pay)

#2개의 PCA 속성을 가진 PCA 객체 생성하고, explained_variance_ratio_ 계산 위해 fit() 호출
scaler = StandardScaler()
df_cols_scaled = scaler.fit_transform(X_features[cols_pay])
pca2 = PCA(n_components=1)
pca2.fit(df_cols_scaled)
PCA2= pca2.transform(df_cols_scaled)

print('PCA Component별 변동성:', pca2.explained_variance_ratio_)
```

```python
X_features['PCA2'] = PCA2
X_features.drop(cols_pay, axis=1,inplace=True)
X_features.head()
```

#### 전체 원본 데이터와 PCA 변환된 데이터 간 랜덤포레스트 예측 성능 비교 

```python
#1. 원본 데이터

import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rcf = RandomForestClassifier(n_estimators=300, random_state=56)

scores = cross_val_score(rcf,X_features,y_target,scoring='accuracy',cv=3)

print('CV=3인 경우의 개별 fold 별 정확도:', scores)
print(f'평균정확도:{np.mean(scores)}')
```

```python
#2. PCA 변환된 데이터

scaler = StandardScaler()
df_scaled= scaler.fit_transform(X_features)

pca_2 = PCA(n_components=7)
df_pca_2 = pca_2.fit_transform(df_scaled)
scores_pca_2 = cross_val_score(rcf, df_pca_2,y_target,scoring='accuracy',cv=3)

print('CV=3인 경우의 PCA 변환된 개별 fold 별 정확도:', scores_pca_2)
print(f'PCA 변환 된 데이터 셋 평균정확도:{np.mean(scores_pca_2)}')
```

### K-means

```python
from sklearn.preprocessing import scale 
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

iris=load_iris()

iris_df=pd.DataFrame(data=iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])

print(iris_df.shape)
iris_df.head(3)
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=0)

kmeans.fit(iris_df)
```

```python
print(kmeans.labels_)
iris_df['cluster'] = kmeans.labels_
```

```python
iris_df['target'] = iris.target

iris_result = iris_df.groupby(['target','cluster'])['sepal_length'].count()
print(iris_result)
```
